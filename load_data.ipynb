```python
# <ticker>,<date>,<open>,<high>,<low>,<close>,<vol>
import csv 

with open('NYSE_20190722.txt', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        print(f"{row['<ticker>']}, {row['<date>']}, {row['<close>']}")
```

```python
from ticker.fetch import fetch_unique_tickers
with open('NYSE_20190722.txt') as csvfile:
    tickers = fetch_unique_tickers(csvfile)
    tickers.sort(key=lambda x: x.ticker)
    for row in tickers:
        print(f"{row.ticker}")
```
